Purpose of this project:
To simulate a basic banking system where users can securely log in, check their balance, deposit, withdraw money, and exit, helping build skills in control flow, input handling, and error management in Python.

```python
# Simple Banking System

# Initialize stored username and password
stored_username = "Aday"
stored_password = "aday1234"

# Initialize account balance


# Allow user 3 login attempts
attempts = 3

# Login loop
# The while loop here keeps running as long as the user has remaining login attempts (attempts > 0). If the correct username and password are entered, it breaks out of the loop; otherwise, it decreases the attempt count and repeats.
while attempts > 0:
    # Prompt user for username and password
    username = input("Enter your username: ")
    password = input("Enter your password: ")

    # Check if entered variables match stored ones
    if username == stored_username and password == stored_password:
        print("\nLogin successful!\n")
        break  # Exit login loop if successful
    else:
        attempts -= 1  # Decrease attempts by 1
        print(f"Incorrect username or password. Attempts left: {attempts}\n")

# If no attempts left, exit the program
if attempts == 0:
  print("Too many failed attempts. Exiting system.")

if attempts != 0:
  balance = 10000.0

# Main menu loop
# The while True: loop runs indefinitely, so the menu keeps reappearing after each operation. The only way to stop it is when the user selects option "4" (Exit), which triggers a break statement to exit the loop and end the program.
  while True:
    print("----- Main Menu -----")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    # Get user's choice
    choice = input("Enter your choice (1-4): ")

    # Check Balance option
    if choice == "1":
        print(f"\nYour current balance is: ${balance:.2f}\n")

    # Deposit Money option
    elif choice == "2":
      # explain how try/except works
        try:
            deposit_amount = float(input("Enter amount to deposit: "))  # Convert input to float
            if deposit_amount > 0:
                balance += deposit_amount  # Add deposit to balance
                print(f"Deposit successful! New balance: ${balance:.2f}\n")
            else:
                print("Deposit amount must be positive.\n")
        except ValueError:
            print("Invalid input. Please enter a number.\n")

    # Withdraw Money option
    elif choice == "3":
        try:
            withdraw_amount = float(input("Enter amount to withdraw: "))  # Convert input to float
            if withdraw_amount > 0:
                if withdraw_amount <= balance:
                    balance -= withdraw_amount  # Subtract withdrawal from balance
                    print(f"Withdrawal successful! New balance: ${balance:.2f}\n")
                else:
                    print("Low balance.\n")
            else:
                print("Withdrawal amount must be positive.\n")
        except ValueError:
            print("Invalid input. Please enter a number.\n")

    # Exit system option
    elif choice == "4":
        print("Thank you for using our banking system. Goodbye!")
        break  # Exit the main menu loop

    # Handle invalid menu choices
    else:
        print("Invalid choice. Please select a valid option (1-4).\n")#
```
