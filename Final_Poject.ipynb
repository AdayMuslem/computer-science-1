```python
# Weather Wardrobe Assistant Program

# Step 1: Define stored username and password
valid_username = "Aday"
valid_password = "aday1234"

# Step 2: Allow the user up to 3 login attempts
login_attempts = 0
max_attempts = 3
logged_in = False  # explain what this is and why set to False

# This code block creates a simple login system that allows a user up to 3 attempts to enter the correct username and password. It checks the input against stored credentials and grants access if they match; otherwise, it shows an error and tracks failed attempts.
while login_attempts < max_attempts:
    username = input("Enter username: ")
    password = input("Enter password: ")

    if username == valid_username and password == valid_password:
        print(f"Welcome, {username}!")
        logged_in = True
        break
    else:
        print("Incorrect username or password. Try again.")
        login_attempts += 1

# Step 3: End the program if login failed after 3 attempts
if not logged_in:
    print("Too many failed attempts. Exiting program.")
    exit()

# Step 4: Initialize variables for weather data
# using None prepares the variables for future assignment in a safe and clear way.
temperature = None
is_raining = None
is_windy = None

# Step 5: Show the main menu until the user chooses to exit
#we use while True loop to continuously display the main menu until the user explicitly chooses to exit
while True:
    print("\nMain Menu:")
    print("1. Enter today’s weather")
    print("2. Get clothing suggestion")
    print("3. Exit")

    choice = input("Enter your choice (1, 2, or 3): ")

    # Step 6: Option 1 - Enter weather data
    #we use while True here to repeatedly prompt the user until valid input is provided for temperature, rain, and wind. This ensures the program only proceeds with correct, clean data.
    if choice == "1":
        while True:
            temp_input = input("Enter the temperature in Celsius: ")
            if temp_input.lstrip('-').isdigit():
                temperature = int(temp_input)
                break
            else:
                print("Please enter a valid number for temperature.")

        while True:
            is_raining = input("Is it raining? (yes or no): ").lower()
            if is_raining in ["yes", "no"]:
                break
            else:
                print("Please answer with 'yes' or 'no'.")

        while True:
            is_windy = input("Is it windy? (yes or no): ").lower()
            if is_windy in ["yes", "no"]:
                break
            else:
                print("Please answer with 'yes' or 'no'.")

        print("Weather data saved!")

    # Step 7: Option 2 - Provide clothing suggestions
    elif choice == "2":
        if temperature is None or is_raining is None or is_windy is None: # explain logic of this line
            print("Please enter today's weather first (Option 1).")
        else:
            print("\nClothing Suggestion:")
            if temperature < 10:
                print("- It's cold. Wear a heavy jacket and warm clothes.")
            elif 10 <= temperature <= 20:
                print("- It's mild. A light jacket or sweater is recommended.")
            else:
                print("- It's warm. T-shirt and shorts are fine.")

            if is_raining == "yes":
                print("- Don't forget an umbrella or a raincoat!")

            if is_windy == "yes":
                print("- It's windy. A windbreaker jacket would help.")

    # Step 8: Option 3 - Exit the program
    elif choice == "3":
        print("Thank you for using the Weather Wardrobe Assistant. Goodbye!")
        break

    # Step 9: Handle invalid menu choices
    else:
        print("Invalid choice. Please select 1, 2, or 3.")
```
